```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
s = np.random.normal(loc=20, scale=5, size=365)
s_weighted = np.zeros(365)
weight = 0.99
for i in range(1,len(s)):
    s_weighted[i] = weight*s[i-1]+(1-weight)*s[i]
```

```python
plt.plot(s)
plt.plot(s_weighted)
```

```python
plt.scatter(list(range(1,366)),s)
plt.scatter(list(range(1,366)),s_weighted)
```
